```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread('../dataset/D2/train/1/7379_right.jpeg')
# image = cv2.imread('../dataset/D4/val/0/24873_right.jpeg')
image = cv2.resize(image, (640, 640))
laplacian = cv2.Laplacian(image, cv2.CV_64F)
edge_image = cv2.convertScaleAbs(laplacian, alpha=5, beta=127)
edge_image = cv2.cvtColor(edge_image, cv2.COLOR_BGR2GRAY)
cv2.imwrite('laplacian.png', edge_image)
print(laplacian.var(), laplacian.shape)

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# fig, ax = plt.subplots(1, 2, figsize=(15,15))
final_img = cv2.equalizeHist(edge_image)
plt.figure(figsize=(6,6), dpi=96)
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.figure(figsize=(6,6), dpi=96)
plt.imshow(edge_image, cmap='gray')
plt.axis('off')
```
